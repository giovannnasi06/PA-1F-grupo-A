Escrever um programa que o usuário digita um número de 1 a 10 e o programa impri a tabuada do número digitado.

```python
palavra = input("Digite um numero de 1 a 10: ")
numero = int(palavra)

for i in range(1, 11):
  print(numero, "x", i, "=", numero * i)
```

Escrever um programa que o usuário digita um número e o programa imprimi até o número digitado.

```python
numero = int(input("Digite um número: "))
for i in range(1, numero + 1):
    print(i)
```

Escrever um programa que o usuário digita um número e o programa imprimi o número digitado até o zero (contagem regressiva).


```python
# Solicita que o usuário digite um número
numero = int(input("Digite um número para a contagem regressiva: "))

print("Iniciando contagem...")

# O loop 'while' continua enquanto o número for maior ou igual a zero
while numero >= 0:
    print(numero)
    numero -= 1  # Subtrai 1 do número a cada repetição (numero = numero - 1)

print("Fim!")
```

Escrever um programa que o usuário digita um número e o programa imprimi os números pares.


```python
numero = int(input("Digite um número inteiro: "))

print(f"Números pares de 0 até {numero}:")

for i in range(numero + 1):
    if i % 2 == 0:
        print(i)
```

Escrever um programa que o usuário digita um número e o programa imprimi os números impares.

```python
numero = int(input("Digite um número: "))

print(f"Números ímpares de 1 até {numero}:")

for num in range(1, numero + 1):
    if num % 2 != 0:
        print(num)
```

Escrever um programa que o usuário digita um número e programa soma os números até o número digitado.

```python
numero = int(input("Digite um número "))
soma = 0
for i in range(1, numero + 1):
    soma += i
print(f"A soma dos números de 1 até {numero} é: {soma}")
```

Escrever um programa que o usuário digita uma palavra e o programa conta as letras da palavra.

```python
palavra = input("Digite uma palavra: ")
quantidade = 0
for letra in palavra:
    quantidade += 1
print(f"A palavra {palavra} tem {quantidade} letras.")
```

Escrever um programa que o usuário digita uma palavra e o programa conta as vogais da palavra.

```python
palavra = input("Digite uma palavra: ").lower()
vogais = "aeiou"
contador = 0

for letra in palavra:
    if letra in vogais:
        contador += 1

print(f"A palavra tem {contador} vogais.")
```

Escrever um programa que o usuário digita uma palavra e o programa imprimi a palavra invertida.

```python
palavra = input("Digite uma palavra: ")
invertida = ""

for i in (palavra):

  invertida = i + invertida
print(invertida)
```

Escrever um programa que o usuário digita 04 números e o programa imprimi o maior.

```python
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))
num3 = float(input("Digite o terceiro número: "))
num4 = float(input("Digite o quarto número: "))

numeros = [num1, num2, num3, num4]

maior = max(numeros)

print(f"O maior número digitado é: {maior}")
```
