```python
#  OPERAÇÕES ARITIMÉTICOS
# Fazer um programa que calcula 02 operações

# Entrada

a = int(input("digite um número:"))
b = int(input("digite um número:"))

# Processamento

some = a + b
subtracao = a - b
multiplicacao = a * b
divisão = a / b
divisao_inteira = a // b
resto = a % b
potencia = a ** b

#Saída

print("")
print("---------------------OPERADORES ARITIMÉTICOS---------------------")
print("")
print("-----------------------------SOMA--------------------------------")
print("")
print (f"O resultado da SOMA de {a} MAIS {b} é igual a : {some}")
print("")
print("--------------------------SUBTRACAO------------------------------")
print("")
print(f"O resultado da SUBTRACAO de {a} MENOS {b} é igual a : {subtracao}")
print("")
print("---------------------------MULTIPLICAÇÃO--------------------------")
print("")
print (f"O resultado da MULTIPLICAÇÃO de {a} POR {b} é igual a : {multiplicacao}")
print("")
print("-----------------------------DIVISÃO-----------------------------")
print("")
print(f"O resultado da DIVISÃO de {a} POR {b} é igual a : {divisão}")
print("")
print("-------------------------DIVISÃO INTEIRA-------------------------")
print("")
print(f"O resultado da DIVISÃO INTEIRA de {a} POR {b} é igual a : {divisao_inteira}")
print("-----------------------------RESTO---------------------------------")
print("")
print(f"O resultado do RESTO de {a} POR {b} é igual a : {resto}")
print("")
print("------------------------------POTENCIA----------------------------")
print("")
print(f"O resultado da POTENCIA de {a} POR {b} é igual a : {potencia}")
print("")
print("-----------------------------------------------------------------")
```
