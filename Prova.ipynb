```python
#Classificação de idade
#Crie um programa que leia a idade de uma pessoa e classifique conforme as regras abaixo:
#Até 17 anos: Menor de idade, de 18 a 59 anos: Adulto, de 60 anos ou mais: Idoso

idade= int(input("Digite sua idade: "))

if idade <= 17:
  print("Você é menor de idade.")

elif idade >= 18 and idade <= 59:
  print("Você é adulto.")

else:
  print("Você é idoso.")
```

```python
#Cálculo de desconto
#Crie um programa que leia o valor de uma compra e aplique o desconto conforme as regras:
#Até R$100: sem desconto, De R$101 a R$500: 10% de desconto,	Acima de R$500: 20% de descontoAo final, mostre o valor final da compra.

preço = float(input("Digite o valor da compra: R$"))
if preço <= 100:
  print("Você não tem desconto")

elif preço >=101 and preço <= 500:
  print("Você tem 10% de desconto")
else:
  print("Você tem 20% de desconto")

```

```python
#triângulo

lado1= int(input("Digite um lado do triângulo: "))
lado2= int(input("Digite um lado do triângulo: "))
lado3= int(input("Digite um lado do triângulo: "))

if (lado1+lado2>lado3) and (lado2+lado3>lado1) and (lado3+lado2>lado1) :
  print("È um triangulo válido")

else:
  print("Não é um triângulo válido")
```

```python
# Crie um programa que leia a energia de um personagem (de 0 a 100) e classifique:
#0: Morto, 1 a 30: Crítico, 31 a 70: Normal,71 a 100: Saudável


# Entrada

Energia = int (input("Digite a energia do seu personagem"))

# Processamento / Saída

if Energia <= 0:
  print (f" Seu personagem tem energia igual a {Energia}. Ele está morto")

elif Energia <=30:
  print (f" Seu personagem tem energia igual a {Energia}. Ele está crítico")

elif Energia <= 70:
  print (f"Seu personagem tem energia igual a {Energia}. Ele está normal")

else:
  print (f"Seu personagem tem energia igual a {Energia}. Ele está saudável")
```
