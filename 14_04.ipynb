```python
#1. Par ou ímpar
#Peça um número ao usuário e informe se ele é par ou ímpar

#Processamento

numero = int(input("Digite um número: "))
if numero % 2 == 0:
    print(f"O número {numero} é PAR.")
else:
    print(f"O número {numero} é ÍMPAR.")
```

```python
#2. Maior de dois números
#Solicite dois números e mostre qual é o maior (ou se são iguais)

#Processamento

num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))

if num1 > num2:
    print(f"O maior número é: {num1}")
elif num2 > num1:
    print(f"O maior número é: {num2}")
else:
    print("Os números são iguais.")
```

```python
#3. Aprovação de aluno Peça a nota de um aluno e informe:
#Aprovado (nota ≥ 7)
#Recuperação (5 ≤ nota < 7)
#Reprovado (nota < 5)

#Processamento

nota = float(input("Digite a nota do aluno: "))

if nota >= 7:
  print("Aprovado")
elif nota >= 5 and nota < 7:
  print("Recuperação")
else:
  print("Reprovado")
```

```python
#4.Classificação de idade
#Receba a idade e classifique:
#Criança (0–12)
#Adolescente (13–17)
#Adulto (18–59)
#Idoso (60+)

#Processamento

idade = int(input("Digite a idade: "))

if idade >= 0 and idade <= 12:
    print("Criança")
elif idade >= 13 and idade <= 17:
    print("Adolescente")
elif idade >= 18 and idade <= 59:
    print("Adulto")
elif idade >= 60:
    print("Idoso")
else:
    print("Idade inválida.")
```

```python
#5.Senha correta
#Defina uma senha fixa no código
#Peça ao usuário para digitar e informe se está correta ou incorreta.

#Processamento

senha_fixa = "eusoulegal23"
senha_digitada = input("Digite a senha: ")
if senha_digitada == senha_fixa:
  print("Senha correta!")
else: print("Senha incorreta.")
```

```python
#6.Positivo, negativo ou zero
#Peça um número e diga se ele é positivo, negativo ou zero.

#Processamento

numero = float(input("Digite um número: "))

if numero > 0:
    print("Positivo")
elif numero < 0:
    print("Negativo")
else:
    print("Zero")
```

```python
#7.	Calculadora simples
#Peça dois números e uma operação (+, -, *, /).
#Use if/elif para calcular o resultado.

#Processamento

num1 = float(input("Digite o primeiro número: "))
operacao = input("Digite a operação (+, -, *, /): ")
num2 = float(input("Digite o segundo número: "))

resultado = None
if operacao == '+':
    resultado = num1 + num2
elif operacao == '-':
    resultado = num1 - num2
elif operacao == '*':
    resultado = num1 * num2
elif operacao == '/':
    if num2 != 0:
        resultado = num1 / num2
    else:
        print("Erro: Divisão por zero!")
else:
    print("Operação inválida.")

if resultado is not None:
    print(f"O resultado é: {resultado}")
```

```python
#8.	Dia da semana
#Peça um número de 1 a 7 e mostre o dia correspondente (ex: 1 = domingo).

#Processamento

dia_numero = int(input("Digite um número de 1 a 7: "))

dias_da_semana = {
    1: "Domingo",
    2: "Segunda-feira",
    3: "Terça-feira",
    4: "Quarta-feira",
    5: "Quinta-feira",
    6: "Sexta-feira",
    7: "Sábado"
}

if dia_numero in dias_da_semana:
    print(f"O dia correspondente é: {dias_da_semana[dia_numero]}")
else:
    print("Número inválido. Digite um número de 1 a 7.")
```

```python
#9.Ano bissexto
#Solicite um ano e informe se ele é bissexto ou não.

#Processamento

ano = int(input("Digite um ano: "))

if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"O ano {ano} é bissexto.")
else:
    print(f"O ano {ano} não é bissexto.")
```

```python
#10.	Desconto em compra
#Peça o valor de uma compra:
#≥ 100: 10% de desconto
#≥ 50 e < 100: 5% de desconto
#< 50: sem desconto

#Processamento

valor_compra = float(input("Digite o valor da compra: "))

desconto = 0
if valor_compra >= 100:
    desconto = 0.10
elif valor_compra >= 50 and valor_compra < 100:
    desconto = 0.05

valor_final = valor_compra * (1 - desconto)
print(f"Valor da compra: R$ {valor_compra:.2f}")
print(f"Desconto aplicado: {desconto * 100:.0f}%")
print(f"Valor final com desconto: R$ {valor_final:.2f}")
```
