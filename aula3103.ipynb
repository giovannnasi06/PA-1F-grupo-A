```python
# Estrutura de decisões

# Escrever um programa que recebe o nome e a idade do usuário e verifica se o usuário é criança, adolescente e adulto

# >= 12 anos é criança
# =< 18 anos é adolescente
# Adulto

# Entrada

nome = str (input ("Digite o seu nome: "))
idade = int (input ("Digite sua idade: "))

# Processamento / Saída

if idade <= 12:
  print (f"{nome}, você tem {idade} anos de idades, você é criança.")

elif idade <= 17:
  print (f"{nome}, você tem {idade} anos de idades, você é adolescente.")

else:

  print (f"{nome}, você tem {idade} anos de idades, você é adulto.")
```

```python
# Em um jogo, um personagem possui uma variável chamada energia. O programa deve: Se a energia for maior que 70 - Forte, se a energia estiver entre 30 e 70 - Cansado se a energia estiver menor que 30 - Fraco.

# Entrada

Energia = int (input("Digite a energia do seu personagem"))

# Processamento / Saída

if Energia <= 30:
  print (f" Seu personagem tem energia igual a {Energia}. Ele está fraco")

elif Energia <= 70:
  print (f"Seu personagem tem energia igual a {Energia}. Ele está cansado")

else:
  print (f"Seu personagem tem energia igual a {Energia}. Ele está forte")
```

```python
# Entrada

lado1 = int(input("Digite o primeiro lado: "))
lado2 = int(input("Digite o segundo lado: "))
lado3 = int(input("Digite o terceiro lado: "))

# Processamento/ Saida

if lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1:
  print("É um triângulo")

if lado1 == lado2 == lado3:
  print("Triângulo equilátero")

elif lado1 != lado2 and lado2 != lado3:
  print("Triângulo escaleno.")

elif lado1 == lado2 and lado1 == lado3:
  print("Triângulo isósceles.")

else:
    print("Não é um triângulo inválido.")
```
